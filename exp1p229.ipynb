# Objective

* The experimental set-up suggests an experiment where the length of an open PVC pipe is used as an independent variable in order to measure resonant frequencies and amplitudes by maintaining a constant diameter across each pipe used and the microphone's relative location within the experimental setup.

* The frequency is how many oscillations are measured per unit time with the units of Hertz (or seconds inverse). The wavelength is defined as the length of one full oscillation - measured in the same units as length (m or cm).

* The length of the pipe which is an independent variable is defined as the distance from one end to the other of the pipe; it is measured in cm.

* The volume is quantified by the amplitude observed on the tablet oscilloscope. It is to be noted that the volume changing makes the amplitude differ as observed. Changing the volume would result in distortions and discontinuities in the graph obtained for frequency vs amplitude.

* The pipes being used have an exact internal diameter (disregarding the thickness) of 4.00cm.

* Resonant frequency is the frequency measured at which a quantity is obtains a high amplitude as a constantly applied/periodic stimulus nears its natural frequency. The volume is controlled respective to the length of the pipe. 

* An antinode is a position where the interference of incoming waves is constructive and results in a maximum amplitude. The pressure antinode is the opposite of a displacement antinode.

* The research question we are trying to answer is how the length of PVC pipes of constant diameter affect the first resonant frequency being measured for a constant input signal.

* My hypothesis is that as the length of the air column increases, the resonant frequency decreases.

# Fitting Function - Lorentzian 

```python
#Acknowledgement : Rob Kiefl, Evan Thomas - PHYS 219; modified script.

def fitting(f,a,ua,fres,amp, gamma, length):
    import numpy as np
    import matplotlib.pyplot as plt
    from scipy.optimize import curve_fit

    x_name = 'Frequency'
    x_units = 'Hz'
    y_name = 'Amplitude'
    y_units = 'V'

    guesses = (amp,fres,gamma,0.0)

    def fit_function(x, Ares,fres,gamma,back):
        return  Ares / np.sqrt(1 + (2*np.pi/(gamma*x))**2 * (x**2-fres**2)**2) +back

    x = f
    y = a
    y_sigma = ua

    ftheory=np.linspace(min(x),max(x),200)

    y_guess = fit_function(ftheory,*guesses)
    plt.errorbar(x,y,yerr=y_sigma,marker='.',linestyle='',label="measured data")
    plt.plot(ftheory,y_guess,marker="",linestyle="-",linewidth=1,color="g",
             label="initial guess")
    plt.xlabel('{} [{}]'.format(x_name,x_units))
    plt.ylabel('{} [{}]'.format(y_name,y_units))
    plt.title(r'Comparison between the data and the intial guess')
    plt.legend(loc='best',numpoints=1)
    print ('\nDisplaying plot 1')
    plt.show()
    plt.figure()

    fit_params,fit_cov = curve_fit(fit_function,x,y,sigma=y_sigma,p0=guesses,
                                   maxfev=10**5)                            

    def chi_square (fit_parameters, x, y, sigma):

        return np.sum((y-fit_function(x, *fit_parameters))**2/sigma**2)

    chi2 = chi_square(fit_params,x,y,y_sigma)
    dof = len(x) - len(fit_params)
    print ("\nGoodness of fit - chi square measure:")
    print ("Chi2 = {}, Chi2/dof = {}\n".format(chi2, chi2/dof))
    fit_cov = fit_cov*dof/chi2
    fit_params_error = np.sqrt(np.diag(fit_cov))
    param_names = ['amplitude','resonant frequency','gamma width','back']
    print ("Fit parameters:")
    for i in range(len(fit_params)):
        print ('{} = {:.3f} +/- {:.3}'.format(param_names[i],
                                              fit_params[i],
                                              fit_params_error[i]))

    x_fitfunc = np.linspace(min(x),max(x),100)
    y_fitfunc = fit_function(x_fitfunc,*fit_params)    

    plt.errorbar(x,y,yerr=y_sigma,marker='.',linestyle='',label="measured data")
    plt.plot(x_fitfunc,y_fitfunc,marker="",linestyle="-",linewidth=2,color="r",
             label=" fit")
    plt.xlabel('{} [{}]'.format(x_name,x_units))
    plt.ylabel('{} [{}]'.format(y_name,y_units))
    plt.title(r'Frequency vs Amplitude plot for length = {} +/- 0.001 m'.format(length))

    plt.legend(loc='best',numpoints=1)
    print ('\nDisplaying plot 2')

    plt.show()

    y_fit=fit_function(x,*fit_params)
    residual = y-y_fit
    normresidual=residual/y_sigma

    hist,bins = np.histogram(normresidual,bins=30)

    fig = plt.figure(figsize=(7,10))
    ax1 = fig.add_subplot(211)
    ax1.errorbar(x,residual,yerr=y_sigma,marker='.',linestyle='',
                 label="residual (y-y_fit)")
    ax1.hlines(0,np.min(x),np.max(x),lw=2,alpha=0.8)
    ax1.set_xlabel('{} [{}]'.format(x_name,x_units))
    ax1.set_ylabel('y-y_fit [{}]'.format(y_units))
    ax1.legend(loc='best',numpoints=1)
    plt.title('Residuals')
    ax2 = fig.add_subplot(212)
    ax2.bar(bins[:-1],hist,width=bins[1]-bins[0])
    ax2.set_ylim(0,1.2*np.max(hist))
    ax2.set_xlabel('(y-y_fit)/y_sigma')
    ax2.set_ylabel('Number of occurences')
    plt.title('Histogram  of normalized residuals')

    within_err=100.*np.sum((residual<=y_sigma)&(residual>=-y_sigma))/len(residual)
    print ("\nResidual information:")
    print ('{:.1f}% of data points agree with fit'.format(within_err))
    ax2.vlines(-1.0,0,np.max(hist)*1.3,lw=2,color='r',linestyles='--')
    ax2.vlines(+1.0,0,np.max(hist)*1.3,lw=2,color='r',linestyles='--',label='+-one sigma')
    ax2.text(0.0,np.max(hist)*1.1,'{:.1f}% of data within one sigma of fit'.format(within_err),
             horizontalalignment='center',verticalalignment='center')
    ax2.legend(loc=(0.35,0.05))

    print ('\nDisplaying plot 3')
    plt.show()
```
